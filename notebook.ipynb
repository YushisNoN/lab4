```python
asm | risc | harv | mc | tick | binary | stream | mem | pstr | prob1 | pipeline |

1) Язык программирования который нужно реализовать Assembler
2) Архитектура процессора risc-iv
3) Память должна быть организована по принципу Гарвардской архитектуры
4) Control Unit: Микрокоманды
5) Точность модели: tick - процессор необходимо моделировать с точностью до такта, процесс моделирования может быть приостановлен на любом такте
6) Представление машинного кода: бинарный. Требуется использовать реальные бинарные файлы, а не текстовые файлы с 0 и 1. + Отладочные вывод в текстовый файл
7) Ввод-вывод: Поток токенов.
8) Ввод-вывод ISA : memory-mapped (порты ввода-вывода отображаются в память и доступ к ним осуществляется штатными командами), отображение портов ввода-вывода в память должно конфигурироваться (можно hardcode-ом).
9) Поддержка строк: Length-prefixed (Pascal string)
10) Largest Palindrome Product
11) pipeline -- конвейерная организация работы процессора. Количество стадий конвейера -- не менее 3.
Необходимо показать влияние конвейера на производительность написанных программ.
В схеме необходимо явно отразить стадии работы конвейера. Пример -- см. в слайдах лекций.
```

Мой Ассемблер:

```python
LOAD Rdest, <address>
    - Загрузка данных из памяти в регистр
        + поддержка прямой адресации
        + поддержка косвенной адресации
        + поддержка относительной адресации

LOADI Rdest, #value
    - Загрузка непосредственного значения в регистр

STORE Rsrc, <address>
    - Сохранение значения из регистра в память
        + поддержка прямой адресации
        + поддержка косвенной адресации
        + поддержка относительной адресации

ADD Rdest, Rsrc1, Rsrc2
    - Сложение двух регистров
        + Rdest = Rsrc1 + Rsrc2

SUB Rdest, Rsrc1, Rsrc2
    - Вычитание
        + Rdest = Rsrc1 - Rsrc2

MUL Rdest, Rsrc1, Rsrc2
    - Умножение

DIV Rdest, Rsrc1, Rsrc2
    - Деление

ADDI Rdest, Rsrc, #value
    - Сложение с константой

SUBI Rdest, Rsrc, #value
    - Вычитание константы

CMP Rsrc1, Rsrc2
    - Сравнение регистров
        + устанавливает флаги (Z, N)

JMP <label>
    - Безусловный переход

JZ <label>
    - Переход если равно нулю

JNZ <label>
    - Переход если не равно

JG <label>
    - Переход если больше

JL <label>
    - Переход если меньше

IN Rdest
    - Ввод значения в регистр

OUT Rsrc
    - Вывод значения из регистра

MOV Rdest, Rsrc
    - Копирование значения между регистрами

NOP
    - Нет операции

HALT
    - Остановка программы
```

```python
Формат инструкций:
R-type:
[opcode][rd][rs1][rs2][000000000000]

I-type:
[opcode][rd][rs1][0000][imm]

J-type:
[opcode][0000][0000][0000][imm]

O-type:
[opcode][0000][0000][0000][000000000000]
```
